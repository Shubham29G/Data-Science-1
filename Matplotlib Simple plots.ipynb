```python
import numpy as np
import matplotlib.pyplot as plt

'''
f(x)=3x−4
f(x) = x^2 + 2x - 15
f(x) = 5(x - 1)(x - 2)(x - 3)
f(x) = e^x
 
f(x)=logx
f(x)=sinx'''

plt.figure(figsize = (12,8))
plt.subplot(3,3,1)
x = np.linspace(0,50,50)
y = 3*x-4
plt.plot(x,y)

plt.subplot(3,3,2)
x = np.linspace(-10,10,30)
y = (x**2)+(2*x)-15
plt.plot(x,y)

plt.subplot(3,3,3)
x  = np.linspace(-5,5,30)
y= 5*(x-1)*(x-2)*(x-3)
plt.plot(x,y)

plt.subplot(3,3,4)
x  = np.linspace(-5,5,30)
y = (np.e)**x
plt.plot(x,y)

plt.subplot(3,3,5)
x  = np.linspace(1,100,30)
y = np.log(x)
plt.plot(x,y)

plt.subplot(3,3,6)
x = np.linspace(-2*np.pi, 2*np.pi,50)
y = np.sin(x)
plt.plot(x,y)

plt.show()
```
